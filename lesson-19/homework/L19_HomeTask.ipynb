```python
import seaborn as sns
titanic = sns.load_dataset('titanic')
titanic
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.histplot(titanic['age'], bins=20, kde=True)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution of Passengers in Titanic')
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.barplot(x='class', y='survived', data=titanic, palette='coolwarm', ci = True)
plt.xlabel('class')
plt.ylabel('Survived')
plt.title('Survived Ratio by class')
plt.show()
```

```python
sns.barplot(x='sex', y='survived', data=titanic)
plt.title('Total Survivors by Gender')
plt.ylabel('Survivors')
plt.show()
```

```python
gender_c = titanic['sex'].value_counts()
plt.figure(figsize=(6, 4))
plt.pie(gender_c, labels=gender_c.index, autopct='%1.1f%%', startangle=130)
plt.title('Ratio of Gender on the Titanic')
plt.axis('equal')
plt.show()
```
