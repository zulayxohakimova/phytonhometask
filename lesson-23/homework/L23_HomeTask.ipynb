```python
df = pd.read_csv('task\\stackoverflow_qa.csv')
df.head(2)
```

1. Find all questions that were created before 2014
2. Find all questions with a score more than 50
3. Find all questions with a score between 50 and 100
4. Find all questions answered by Scott Boston
5. Find all questions answered by the following 5 users
6. Find all questions that were created between March, 2014 and October 2014 that were answered by Unutbu and have score less than 5.
7. Find all questions that have score between 5 and 10 or have a view count of greater than 10,000
8. Find all questions that are not answered by Scott Boston

```python
#1
df['creationdate'] = pd.to_datetime(df['creationdate'], errors='coerce')

df['creationyear'] = df['creationdate'].dt.year

crit1 = df['creationyear'] < 2014
dta = df[crit1]
print(dta)
```

```python
#2
crit2 = df['score'] > 50
df[crit2]
```

```python
#3
crit3 = (df['score'] > 50) & (df['score'] < 100)
df[crit3]
```

```python
#4
crit4 = df['ans_name'] == 'Scott Boston'
df[crit4]
```

```python
#5
crit = df['ans_name'].head(6).dropna()
crit5 = df['ans_name'].isin(crit)
df[crit5]
```

```python
#6 Find all questions that were created between March, 2014 and October 2014 that were answered by Unutbu and have score less than 5.
crit1 = (df['creationdate'].dt.year == 2014) & (df['creationdate'].dt.month.isin([3, 4, 5, 6, 7, 8, 9, 10]))
crit2 = df['ans_name'] == 'unutbu'
crit3 = df['score'] < 5
crit_final = crit1 & crit2 & crit3
df[crit_final]
```

```python
#7  Find all questions that have score between 5 and 10 or have a view count of greater than 10,000
crit1 = df['score'].between(5, 10)
crit2 = df['viewcount'] > 10000
crit_final = crit1 & crit2
df[crit_final]
```

```python
#8 Find all questions that are not answered by Scott Boston
crit8 = df['ans_name'] != 'Scott Boston'
df[crit8]
```
