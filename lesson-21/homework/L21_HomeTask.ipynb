```python
# Puzzle 1
# Write a query which will extract the previous value from the currentQuota for each row.

import pandas as pd

data = [
    (275, 2005, 367000.00),
    (275, 2005, 556000.00),
    (275, 2006, 502000.00),
    (275, 2006, 550000.00),
    (275, 2006, 1429000.00),
    (275, 2006, 1324000.00)
]

columns = ['BusinessEntityID', 'SalesYear', 'CurrentQuota']
df = pd.DataFrame(data, columns=columns)

print(df)

df['lagCurrentData'] = df['CurrentQuota'].shift(1)
df.fillna(0)
```

```python
# Puzzle 2
# Write a query which will extract the next value from the currentQuota for each row.
data = [
    (275, 2005, 367000.00),
    (275, 2005, 556000.00),
    (275, 2006, 502000.00),
    (275, 2006, 550000.00),
    (275, 2006, 1429000.00),
    (275, 2006, 1324000.00)
]
columns = ['BusinessEntityID', 'SalesYear', 'CurrentQuota']
df = pd.DataFrame(data, columns=columns)

df['NextCurrentData'] = df['CurrentQuota'].shift(-1)
df.fillna(0)
```
